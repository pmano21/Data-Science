```python
import math
def isprime(n):
  if n<=1:
    return 0
  if n==2:
    return 1
  if n%2==0 and n>2:
    return 0
  div=int(math.sqrt(n))
  for i in range(3,div+1,2):
    if n%i==0:
      return 0
  return 1

a=np.array([2,5,6,7,8,9,1,4,11,13])
b=[]
for ele in a:
  if isprime(ele)==1:
    b.append(ele)
print(b)
    
  
```
