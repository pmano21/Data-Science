```python
import pandas as pd
path="/content/drive/MyDrive/Salary_Data.csv"
df=pd.read_csv(path)  
```

```python
import numpy as np
```

```python
df.head()
```

```python
x=df.loc[:,['YearsExperience']]
x.head()
```

```python
y=df.loc[:,['Salary']]
y.head()
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.xlabel('ExperienceYears')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn import linear_model
```

```python
lm=linear_model.LinearRegression()
model1=lm.fit(x_train,y_train)
```

```python
model1.coef_
```

```python
model1.intercept_
```

```python
type(model1)
```

```python
y_pred=model1.coef_*x_train + model1.intercept_
```

```python
plt.plot(x_train,y_pred,color="g") 
plt.xlabel('x')                    
plt.ylabel('y')
plt.scatter(x_train,y_train,color='m',marker='o',s=40) 
plt.show()
```

```python
test_pred=model1.predict(x_test)
test_pred
```

```python
 cost=np.sqrt(np.mean(np.sum(np.square(y_test-test_pred))))
 cost
```
