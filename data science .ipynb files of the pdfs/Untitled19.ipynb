```python
import pikepdf
from tqdm import tqdm
passwords=[line.strip()for line in open("pswd.txt.txt")]
for password in tqdm(passwords,"Decryprting PDF"):
    try:
        with pikepdf.open("check1.pdf",password=password)as pdf:
            print("\n[+]Password found:",password)
            break
    except pikepdf._qpdf.PasswordError as e:
        #wrong password,just continue in the loop
        continue
```
