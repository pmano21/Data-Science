```python
import math
def prime(x):

  for i in range(2,int(math.sqrt(x))+1):
    if x%i==0:
      return False
  return True
prime(3)
```

```python
def isprime(num):
    for n in range(2,int(num**0.5)+1):
        if num%n==0:
            return False
    return True
isprime(4)
```
